***Exploratory Data Analysis***

```python
df['age'].value_counts()
```

```python
a = df['age_range']=pd.cut(df['age'],bins= [0,18,30,40,60,100])
print(a)
```

```python
#making a hist plot for age count
sns.histplot(data = df, kde = True)
plt.title('Age Count Plot', fontsize=20)
plt.xlabel("age", fontsize=20)
plt.ylabel("count", fontsize=20)
```

```python
# making the subscriber aagroup countplot
sns.countplot(x="age_range",hue = "y", data=df)
plt.title('Subscribed by Age Group', fontsize=20)
plt.xlabel("age", fontsize=20)
plt.ylabel("count", fontsize=20)
plt.show()
```

```python
# making a kde plot for age and campaign
plt.figure(figsize=(15,15))
sns.kdeplot(
   data=df, x="age", hue="campaign",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0
)
plt.title('Age Group Of People Who Attended The Marketing Campaign', fontsize=20)
```

```python
# making a graph for subscriber by different kind of jobs
sns.countplot(x="age_range",hue = "job", data=df)
plt.title('People Subscribed By Different Kind Of Jobs', fontsize=15)
plt.xlabel("age", fontsize=10)
plt.ylabel("count", fontsize=10)
plt.show()
```

```python
sns.barplot(data=df, x="age", y="marital", hue="marital")
```

```python
# making a plot for age and loan
plt.figure(figsize=(30,30))
sns.countplot(x="age", hue="loan", data=df)
plt.title('People who got loan in Different age group', fontsize=15)
plt.xlabel("age", fontsize=10)
plt.ylabel("loan", fontsize=10)
plt.show()
```

```python
#making a
plt.figure(figsize=(30,30))
sns.histplot(x="age", hue="marital", data=df)
plt.title('Martial Status Of The People', fontsize=15)
plt.xlabel("age", fontsize=10)
plt.ylabel("marital", fontsize=10)
plt.show()
```

***Correlation Heatmap***

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(),annot=True, cmap=sns.color_palette("Spectral", as_cmap=True), alpha=0.9)
plt.show()
```
